```python
import requests
from bs4 import BeautifulSoup
import csv

url = 'https://webscraper.io/test-sites/e-commerce/allinone/computers/laptops'

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'
}  # Helps avoid basic bot detection

response = requests.get(url, headers=headers)
if response.status_code != 200:
    print(f"Failed to fetch page: {response.status_code}")
    exit()

soup = BeautifulSoup(response.text, 'html.parser')

# Find product containers – usually div.thumbnail or div.col-sm-4.col-lg-4.col-md-4
products = soup.find_all('div', class_='thumbnail')

print(f"Found {len(products)} product containers")

data = []
category = 'Laptops'

for i, product in enumerate(products, 1):
    # Name
    name_tag = product.find('a', class_='title')
    name = name_tag.text.strip() if name_tag else "N/A"

    # Price – try modern variations
    price_tag = None
    
    # Option 1: Current common pattern (float-end pull-right price)
    price_tag = product.find('h4', class_='float-end pull-right price')
    
    # Option 2: Old pattern that sometimes still appears
    if not price_tag:
        price_tag = product.find('h4', class_='pull-right price')
    
    # Option 3: Fallback – any h4 with $ in text (very safe for this site)
    if not price_tag:
        price_tag = product.find('h4', string=lambda t: t and '$' in t)
    
    # Option 4: span or div with price class
    if not price_tag:
        price_tag = product.find(lambda tag: tag.name in ['h4', 'span', 'div'] and 'price' in tag.get('class', []))

    price = price_tag.text.strip() if price_tag else "N/A"

    # Ratings: count filled stars
    stars = product.find_all('span', class_='ws-icon ws-icon-star')
    ratings = len(stars)  # usually 0–5

    # Reviews
    review_tag = product.find('p', class_='pull-right')
    reviews = review_tag.text.strip().split()[0] if review_tag else "0"

    data.append([name, price, category, ratings, reviews])

    # Debug: show what we got for first few
    if i <= 3:
        print(f"Product {i}: {name} | Price: {price} | Stars: {ratings} | Reviews: {reviews}")

# Save to CSV even if some fields missing
with open('laptops.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(['Product Name', 'Price', 'Category', 'Ratings (stars)', 'Number of Reviews'])
    writer.writerows(data)

print(f"\nScraped {len(data)} products → saved to laptops.csv")
```

```python
# Data Cleaning
# Clean price data - extract numeric values
def clean_price(price_str):
    if pd.isna(price_str) or price_str == "N/A":
        return np.nan
    # Remove $ and commas, convert to float
    price_clean = re.sub(r'[^\d.]', '', str(price_str))
    try:
        return float(price_clean)
    except:
        return np.nan

# Clean reviews data - extract numeric values
def clean_reviews(review_str):
    if pd.isna(review_str) or review_str == "N/A":
        return 0
    # Extract number from review string
    numbers = re.findall(r'\d+', str(review_str))
    return int(numbers[0]) if numbers else 0

# Apply cleaning functions
df['Price_Clean'] = df['Price'].apply(clean_price)
df['Reviews_Clean'] = df['Number of Reviews'].apply(clean_reviews)

# Handle missing values
df_clean = df.copy()
df_clean['Price_Clean'].fillna(df_clean['Price_Clean'].median(), inplace=True)
df_clean['Product Name'].fillna('Unknown', inplace=True)

# Remove duplicates
df_clean.drop_duplicates(subset=['Product Name', 'Price_Clean'], keep='first', inplace=True)

# Standardize text casing
df_clean['Product Name'] = df_clean['Product Name'].str.title()
df_clean['Category'] = df_clean['Category'].str.title()

print("Cleaned dataset shape:", df_clean.shape)
print("\nMissing values after cleaning:")
print(df_clean.isnull().sum())

print("\nSample of cleaned data:")
print(df_clean[['Product Name', 'Price_Clean', 'Category', 'Ratings (stars)', 'Reviews_Clean']].head())
```

```python
# Database Setup with SQLAlchemy
from sqlalchemy import create_engine, Column, Integer, String, Float
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import sessionmaker

# Create SQLite database
engine = create_engine('sqlite:///laptop_data.db', echo=False)
Base = declarative_base()

# Define the Laptop model
class Laptop(Base):
    __tablename__ = 'laptops'
    
    id = Column(Integer, primary_key=True)
    product_name = Column(String(255))
    price = Column(Float)
    category = Column(String(100))
    ratings = Column(Integer)
    reviews = Column(Integer)
    brand = Column(String(100))

# Create the table
Base.metadata.create_all(engine)

# Create a session
Session = sessionmaker(bind=engine)
session = Session()

# Insert cleaned data into database
for _, row in df_clean.iterrows():
    laptop = Laptop(
        product_name=row['Product Name'],
        price=row['Price_Clean'],
        category=row['Category'],
        ratings=row['Ratings (stars)'],
        reviews=row['Reviews_Clean'],
        brand=row['Brand']
    )
    session.add(laptop)

# Commit the transaction
session.commit()

print("Data successfully stored in SQLite database!")
print(f"Total records inserted: {session.query(Laptop).count()}")

# Verify data retrieval
sample_data = session.query(Laptop).limit(5).all()
print("\nSample data from database:")
for laptop in sample_data:
    print(f"{laptop.product_name} - ${laptop.price} - {laptop.brand} - {laptop.ratings} stars")

session.close()
```

```python
# Unsupervised Learning - K-means Clustering
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Retrieve data from database for clustering
session = Session()
laptops_data = session.query(Laptop).all()
session.close()

# Prepare features for clustering
features = []
for laptop in laptops_data:
    features.append([
        laptop.price,
        laptop.ratings,
        laptop.reviews
    ])

X = np.array(features)

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Find optimal number of clusters using Elbow Method and Silhouette Score
inertias = []
silhouette_scores = []
k_range = range(2, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_scaled, kmeans.labels_))

# Plot Elbow Method and Silhouette Score
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

ax1.plot(k_range, inertias, 'bo-')
ax1.set_xlabel('Number of clusters (k)')
ax1.set_ylabel('Inertia')
ax1.set_title('Elbow Method')
ax1.grid(True)

ax2.plot(k_range, silhouette_scores, 'ro-')
ax2.set_xlabel('Number of clusters (k)')
ax2.set_ylabel('Silhouette Score')
ax2.set_title('Silhouette Score Analysis')
ax2.grid(True)

plt.tight_layout()
plt.show()

# Choose optimal k (highest silhouette score)
optimal_k = k_range[np.argmax(silhouette_scores)]
print(f"Optimal number of clusters: {optimal_k}")
print(f"Best silhouette score: {max(silhouette_scores):.3f}")
```

```python
# Apply K-means with optimal clusters and analyze results
kmeans_final = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
cluster_labels = kmeans_final.fit_predict(X_scaled)

# Add cluster labels to dataframe
df_clean['Cluster'] = cluster_labels

# Analyze clusters
cluster_analysis = df_clean.groupby('Cluster').agg({
    'Price_Clean': ['mean', 'std', 'count'],
    'Ratings (stars)': 'mean',
    'Reviews_Clean': 'mean',
    'Brand': lambda x: x.mode().iloc[0] if len(x.mode()) > 0 else 'Mixed'
}).round(2)

print("Cluster Analysis:")
print(cluster_analysis)

# Visualize clusters
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Price vs Ratings colored by cluster
scatter = axes[0, 0].scatter(df_clean['Ratings (stars)'], df_clean['Price_Clean'], 
                           c=cluster_labels, cmap='viridis', alpha=0.6)
axes[0, 0].set_xlabel('Rating (Stars)')
axes[0, 0].set_ylabel('Price ($)')
axes[0, 0].set_title('Price vs Ratings by Cluster')
plt.colorbar(scatter, ax=axes[0, 0])

# Price vs Reviews colored by cluster
scatter2 = axes[0, 1].scatter(df_clean['Reviews_Clean'], df_clean['Price_Clean'], 
                            c=cluster_labels, cmap='viridis', alpha=0.6)
axes[0, 1].set_xlabel('Number of Reviews')
axes[0, 1].set_ylabel('Price ($)')
axes[0, 1].set_title('Price vs Reviews by Cluster')
plt.colorbar(scatter2, ax=axes[0, 1])

# Cluster distribution
cluster_counts = df_clean['Cluster'].value_counts().sort_index()
axes[1, 0].bar(cluster_counts.index, cluster_counts.values, color='skyblue')
axes[1, 0].set_xlabel('Cluster')
axes[1, 0].set_ylabel('Number of Products')
axes[1, 0].set_title('Products per Cluster')

# Average price by cluster
avg_price_cluster = df_clean.groupby('Cluster')['Price_Clean'].mean()
axes[1, 1].bar(avg_price_cluster.index, avg_price_cluster.values, color='lightcoral')
axes[1, 1].set_xlabel('Cluster')
axes[1, 1].set_ylabel('Average Price ($)')
axes[1, 1].set_title('Average Price by Cluster')

plt.tight_layout()
plt.show()

# Save clustered data to new CSV
df_clean.to_csv('laptops_clustered.csv', index=False)
print(f"\nClustered data saved to 'laptops_clustered.csv'")
```

```python
# Supervised Learning - Classification Models
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns

# Prepare data for classification
# We'll predict Brand based on Price, Ratings, and Reviews
X = df_clean[['Price_Clean', 'Ratings (stars)', 'Reviews_Clean']].values
y = df_clean['Brand'].values

# Encode target variable
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

# Standardize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print(f"Training set shape: {X_train.shape}")
print(f"Test set shape: {X_test.shape}")
print(f"Number of classes (brands): {len(le.classes_)}")
print(f"Brand classes: {le.classes_}")
```

```python
# Initialize and train all models
models = {
    'Logistic Regression': LogisticRegression(random_state=42, max_iter=1000),
    'SVM': SVC(random_state=42),
    'k-NN': KNeighborsClassifier(),
    'Random Forest': RandomForestClassifier(random_state=42),
    'XGBoost': XGBClassifier(random_state=42, eval_metric='mlogloss')
}

# Train and evaluate each model
results = {}
predictions = {}

for name, model in models.items():
    print(f"\nTraining {name}...")
    
    # Train the model
    if name in ['Logistic Regression', 'SVM', 'k-NN']:
        model.fit(X_train_scaled, y_train)
        y_pred = model.predict(X_test_scaled)
    else:
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
    
    # Store predictions
    predictions[name] = y_pred
    
    # Calculate metrics
    accuracy = accuracy_score(y_test, y_pred)
    results[name] = {
        'accuracy': accuracy,
        'model': model
    }
    
    print(f"{name} Accuracy: {accuracy:.4f}")

# Display results comparison
print("\n=== MODEL COMPARISON ===")
for name, result in sorted(results.items(), key=lambda x: x[1]['accuracy'], reverse=True):
    print(f"{name}: {result['accuracy']:.4f}")

# Find best model
best_model_name = max(results.keys(), key=lambda k: results[k]['accuracy'])
best_model = results[best_model_name]['model']
print(f"\nBest performing model: {best_model_name} with accuracy {results[best_model_name]['accuracy']:.4f}")
```

```python
# Detailed evaluation of the best model
print(f"\n=== DETAILED EVALUATION FOR {best_model_name} ===")

# Classification report
y_pred_best = predictions[best_model_name]
print("\nClassification Report:")
print(classification_report(y_test, y_pred_best, target_names=le.classes_))

# Confusion Matrix
plt.figure(figsize=(10, 8))
cm = confusion_matrix(y_test, y_pred_best)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=le.classes_, yticklabels=le.classes_)
plt.title(f'Confusion Matrix - {best_model_name}')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

# Feature importance (for tree-based models)
if best_model_name in ['Random Forest', 'XGBoost']:
    feature_names = ['Price', 'Ratings', 'Reviews']
    importances = best_model.feature_importances_
    
    plt.figure(figsize=(8, 6))
    plt.bar(feature_names, importances, color='skyblue')
    plt.title(f'Feature Importance - {best_model_name}')
    plt.xlabel('Features')
    plt.ylabel('Importance')
    plt.show()
    
    for name, importance in zip(feature_names, importances):
        print(f"{name}: {importance:.4f}")
```

```python
# Hyperparameter Tuning for the Best Model
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
import warnings
warnings.filterwarnings('ignore')

print(f"Performing hyperparameter tuning for {best_model_name}...")

# Define parameter grids for different models
param_grids = {
    'Logistic Regression': {
        'C': [0.1, 1, 10, 100],
        'penalty': ['l1', 'l2'],
        'solver': ['liblinear', 'saga']
    },
    'SVM': {
        'C': [0.1, 1, 10, 100],
        'kernel': ['linear', 'rbf', 'poly'],
        'gamma': ['scale', 'auto', 0.001, 0.01, 0.1, 1]
    },
    'k-NN': {
        'n_neighbors': [3, 5, 7, 9, 11],
        'weights': ['uniform', 'distance'],
        'metric': ['euclidean', 'manhattan']
    },
    'Random Forest': {
        'n_estimators': [50, 100, 200],
        'max_depth': [None, 10, 20, 30],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4]
    },
    'XGBoost': {
        'n_estimators': [50, 100, 200],
        'max_depth': [3, 6, 9],
        'learning_rate': [0.01, 0.1, 0.2],
        'subsample': [0.8, 0.9, 1.0]
    }
}

# Get the parameter grid for the best model
param_grid = param_grids.get(best_model_name, {})

if param_grid:
    # Create a fresh instance of the best model
    if best_model_name == 'Logistic Regression':
        model_tune = LogisticRegression(random_state=42, max_iter=1000)
        X_tune = X_train_scaled
        y_tune = y_train
    elif best_model_name == 'SVM':
        model_tune = SVC(random_state=42)
        X_tune = X_train_scaled
        y_tune = y_train
    elif best_model_name == 'k-NN':
        model_tune = KNeighborsClassifier()
        X_tune = X_train_scaled
        y_tune = y_train
    elif best_model_name == 'Random Forest':
        model_tune = RandomForestClassifier(random_state=42)
        X_tune = X_train
        y_tune = y_train
    elif best_model_name == 'XGBoost':
        model_tune = XGBClassifier(random_state=42, eval_metric='mlogloss')
        X_tune = X_train
        y_tune = y_train
    
    # Perform Grid Search
    print(f"Running Grid Search for {best_model_name}...")
    grid_search = GridSearchCV(
        estimator=model_tune,
        param_grid=param_grid,
        cv=5,
        scoring='accuracy',
        n_jobs=-1,
        verbose=1
    )
    
    grid_search.fit(X_tune, y_tune)
    
    print(f"\nBest parameters found: {grid_search.best_params_}")
    print(f"Best cross-validation accuracy: {grid_search.best_score_:.4f}")
    
    # Evaluate the tuned model on test set
    if best_model_name in ['Logistic Regression', 'SVM', 'k-NN']:
        y_pred_tuned = grid_search.predict(X_test_scaled)
    else:
        y_pred_tuned = grid_search.predict(X_test)
    
    tuned_accuracy = accuracy_score(y_test, y_pred_tuned)
    original_accuracy = results[best_model_name]['accuracy']
    
    print(f"\nOriginal test accuracy: {original_accuracy:.4f}")
    print(f"Tuned test accuracy: {tuned_accuracy:.4f}")
    print(f"Improvement: {tuned_accuracy - original_accuracy:.4f}")
    
    # Save the best tuned model
    best_tuned_model = grid_search.best_estimator_
    
else:
    print(f"No parameter grid defined for {best_model_name}")
```
